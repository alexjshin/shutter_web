```python
# I want to use the skeleton joints and use the plotly package to create a visualiziation of the skeleton
# I also want to use the skeleton sin and cos joints to help create this visualiziation of the skeleton (rotation)
# create a trace for the skeleton data
for index, row in comb_df.iterrows():
    
    skeleton_trace = go.Scatter3d(
        x=row[skel_x_joints],
        y=row[skel_y_joints],
        z=row[skel_z_joints],
        mode = 'markers',
        marker = dict(
            size = 3,
            color = 'blue',
            opacity = 0.7),
        name = 'skeleton'
    )

    # create a layout for the plot
    skeleton_layout = go.Layout(
        title = 'Skeleton Visualizations over Time',
        scene = dict(
            xaxis = dict(title = 'x', range = [0, 4], tickmode = 'linear', tick0 = 0, dtick = 1),
            yaxis = dict(title = 'y', range = [-4, 4], tickmode = 'linear', tick0 = -4, dtick = 1),
            zaxis = dict(title = 'z', range = [-1, 2], tickmode = 'linear', tick0 = -1, dtick = 1),
            aspectmode = 'cube'
        ),
        width = 700,
        height = 800,
        annotations = [
        dict(
            x = 0.5,
            y = 1.05,
            xref = 'paper',
            yref = 'paper',
            text = 'Time: ' + str(comb_df['resampled_timestamp'].iloc[index]),
            showarrow = False,
            font = dict(
                size = 15,
                color = 'black'
            )
            )
        ]
    )

    # create a figure and add the skeleton trace and layout
    skeleton_fig = go.Figure(data=[skeleton_trace], layout=skeleton_layout) 

    # show the figure
    skeleton_fig.show()

    print(index)
    print(comb_df['resampled_timestamp'].iloc[index])

    if (index+1 < 10):
        filename = '0' + str(index+1) + "_skeleton" + '.json'
    else:
        filename = str(index+1) + "_skeleton" + '.json'

    pio.write_json(skeleton_fig, file=filename)
```

```python
#same visualization as above except with line tracing
for index, row in comb_df.iterrows():
    
    skeleton_trace = go.Scatter3d(
        x=row[skel_x_joints],
        y=row[skel_y_joints],
        z=row[skel_z_joints],
        mode = 'markers',
        marker = dict(
            size = 3,
            color = 'blue',
            opacity = 0.7),
        name = 'skeleton'
    )
    
    skeleton_line = go.Scatter3d(
        x=row[skel_x_joints],
        y=row[skel_y_joints],
        z=row[skel_z_joints],
        mode = 'lines',
        line = dict(
            color = 'blue',
            width = 2,
            cauto = False
        ),
        name = 'skeleton_line'
    )

    # create a layout for the plot
    skeleton_layout = go.Layout(
        title = 'Skeleton Visualizations over Time',
        scene = dict(
            xaxis = dict(title = 'x', range = [0, 4], tickmode = 'linear', tick0 = 0, dtick = 1),
            yaxis = dict(title = 'y', range = [-4, 4], tickmode = 'linear', tick0 = -4, dtick = 1),
            zaxis = dict(title = 'z', range = [-1, 2], tickmode = 'linear', tick0 = -1, dtick = 1),
            aspectmode = 'cube'
        ),
        width = 700,
        height = 800,
        annotations = [
        dict(
            x = 0.5,
            y = 1.05,
            xref = 'paper',
            yref = 'paper',
            text = 'Time: ' + str(comb_df['resampled_timestamp'].iloc[index]),
            showarrow = False,
            font = dict(
                size = 15,
                color = 'black'
            )
            )
        ]
    )

    # create a figure and add the skeleton trace and layout
    skeleton_fig = go.Figure(data=[skeleton_trace, skeleton_line], layout=skeleton_layout)  

    # show the figure
    skeleton_fig.show()

    print(index)
    print(comb_df['resampled_timestamp'].iloc[index])

    if (index+1 < 10):
        filename = '0' + str(index+1) + "_skeleton_connected" + '.json'
    else:
        filename = str(index+1) + "_skeleton_connected" + '.json'

    pio.write_json(skeleton_fig, file=filename)
```
